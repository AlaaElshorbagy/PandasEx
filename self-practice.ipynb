## Changing multiple rows data - apply/lambda/applymap/map/replace

```python
#say we want to change all emails to lower case 

people_df["EMAIL"] = people_df['EMAIL'].str.lower()
people_df
```

###### lambda function 
 usualy it is used for simple functions

#### Method 1: apply 

Used for calling a function on our values -- Can be work on dataframe or series object -- it is behavior may depends on the type of the object. 

#####  a- apply on a series object

```python
#lets say we want the length of all the email address
#that is we want to apply the len() function to each values in our series
people_df['EMAIL'].apply(len)
```

```python
#apply more complicated functions
def update_email(email):
    return email.lower()
people_df["EMAIL"].apply(update_email) # that is not assigned to our coulumn 
```

```python
people_df
```

```python
#To update the dataframe itself we need to assign that value to the column 
people_df["EMAIL"] = people_df["EMAIL"].apply(update_email)
people_df
```

#### b- apply on dataframe
It applies on each series in that dataframe -- could be columns (default setting) or rows.

```python
people_df.apply(len) 
#it applies on each series (aka column) in the data frame, i.e. it measures the number of elements in each column
```

```python
people_df.apply(len, axis="columns")  # it I want to counts the elemnets in the rows(aka no of columns) in each row
```

```python
#get the min of each column
people_df.apply(pd.Series.min) #series objects has a min function that can be applied
# for strings it returns the first in alphabetic order
```

#### Method 2: applymap
Only work on dataframe -- used to apply function on each value inside the dataframe

```python
people_df.applymap(len) 
```

```python
people_df.applymap(str.lower) #works only if all values in the dataframe are strings
#recall non of this methods change the actual column unless we assign it
```

#### Method 3: map

only works on series - used to substitute each value on a series with another value  
change some of the values in a column and assign NaN to the remaining item

```python
#say we want to change some of the values in a column and assign NaN to the remaining item
# we do that by passing a dictionary to the .map method
people_df['first'].map({'Islam':"smsm", 'Alaa': 'lolo'}) #recall non of this methods change the actual column unless we assign it
```

```python
#say we want to convert the values of that column to a True False (Boolean)
df['Hobbyist'].map({'Yes':True, "No":False}) 
```

#### mehtod 4: replace
change some of the values in a column and keep the remaining items

```python
#say we want to change some of the values in a column and keep the remaining items
# we do that by passing a dictionary to the .replace method
people_df['first'].replace({'Islam':"smsm", 'Alaa': 'lolo'}) #recall non of this methods change the actual column unless we assign it
```

### Double filtering

#### & (and) | (or)  operator

```python
filt = (df['Country'] == 'United Kingdom') & (df['Student'] == 'No')
df[filt]
```

```python
filt = (df['Country'] == 'United Kingdom') | (df['Country'] == 'Canada') #or operator
df[filt]
```

```python
# First way: boring and long 
filt = (df["ConvertedComp"]>70000) & ((df['Country']== 'India') |  (df['Country']== 'United Kingdom'))
df.loc[filt,['Country', 'LanguageWorkedWith', 'ConvertedComp']]
```

```python
#secound way: Faster and smarter
countries = ['United States', 'India', 'United Kingdom', 'Germany', 'Canada']
filt = df['Country'].isin(countries) #pick the rows where the 'Country' is in the list 'countries'
df.loc[filt,'Country']
```

### use .loc to filter for specific column/s

```python
df.loc[filt,['Hobbyist']]
```

```python
df.loc[filt,'Hobbyist']
```
